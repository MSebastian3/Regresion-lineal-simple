## **Limpieza**

### host_acceptance_rate

```python
# Eliminar los símbolos de porcentaje
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace('%', '')

#filtro sin 'sin registro'
filtro_sin_registro = df.loc[df['host_acceptance_rate'] != 'sin registro']

# Convertir la columna a tipo float
filtro_sin_registro['host_acceptance_rate'] = filtro_sin_registro['host_acceptance_rate'].astype(float)

# Calcular el promedio
promedio = filtro_sin_registro['host_acceptance_rate'].mean()
promedio
```

```python
# Eliminar los símbolos de porcentaje
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace('sin registro', "91.4558611361587")

# Convertir la columna a tipo float
df['host_acceptance_rate'] = df['host_acceptance_rate'].astype(float)
```

## **Filtro1**

### host_acceptance_rate vs host_response_rate

```python
#Imprimimos el scatter plot entre la variable dependiente (total) e independiente (alchool) 
#para observar el comportamiento en su dispersión 
from turtle import color


#x independiente. y dependiente

sns.scatterplot(x='host_acceptance_rate', y='host_response_rate', color="blue", data=filtro1)
#sns.scatterplot(X='speeding', y='total', color="red", data=df)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
#Vars_Indep= df[['alcohol', 'speeding']]
Vars_Indep= df[['host_acceptance_rate']]
Var_Dep= df['host_response_rate']
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

**Modelo matemático: y= 0.31816203x + 65.44073745167825**

---




```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
model.score(Vars_Indep,Var_Dep)
```

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
